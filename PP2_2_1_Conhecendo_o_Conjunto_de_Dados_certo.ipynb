```python
# Lendo o arquivo CSV baixado
import os

# Monta o caminho completo até o arquivo CSV
csv_path = os.path.join(path, "covtype.csv")

# 1.Carrega a base de dados
df = pd.read_csv(csv_path)

# 2.Mostra as 10 primeiras linhas
df.head(10)

```

```python
# 3. Mostra os 10 últimos exemplos da base de dados
df.tail(10)
```

```python

#4. Mostra quantos exemplos
print("Número de exemplos:", df.shape[0])
```

```python

#  5. Criando tabela automática com o tipo de dado interpretado
atributos = []

for col in df.columns:
    unique_vals = df[col].nunique()  # conta quantos valores únicos existem

    if col == "Cover_Type":
        tipo = "Alvo (1 a 7)"
    elif unique_vals == 2:
        tipo = "Binário (0/1)"
    elif df[col].dtype in ['int64', 'float64']:
        tipo = "Numérico (inteiro ou real)"
    else:
        tipo = str(df[col].dtype)  # caso seja outro tipo

    atributos.append([col, tipo, unique_vals])

atributos_df = pd.DataFrame(atributos, columns=["Atributo", "Tipo", "Valores Únicos"])
atributos_df

```

```python
# 6. Total de valores nulos na base inteira
total_vazios = df.isnull().sum().sum()
print("Total de valores vazios na base:", total_vazios)
```

7.os atributos “Soil_Type1” a “Soil_Type40” representam tipos de solo encontrados nas áreas de floresta do estado do Colorado (EUA), conforme mapeados pelo U.S. Geological Survey (USGS).

Cada uma dessas 40 colunas é uma variável binária (0 ou 1) que indica se aquele tipo de solo está presente (1) ou não (0) no ponto observado.

7.1 É qualitativo (categórico), pois representa categorias de solo, e não valores numéricos com significado quantitativo
7.2 A codificação usada é chamada de One-Hot Encoding (Codificação binária ou dummy).

Cada tipo de solo é representado por uma coluna binária independente.
Somente uma dessas colunas tem valor 1 para cada exemplo — o restante permanece 0.

3.1 a tarefa é uma classificação multiclasse supervisionada

Supervisionada porque temos dados de entrada e saída conhecida, e multiclasse → porque existem mais de 2 classes possíveis (7 tipos de floresta).

**Acurácia (Accuracy)**:Percentual de acertos sobre o total de exemplos.

**Precisão (Precision) por classe**: Fração de exemplos classificados corretamente entre todos os exemplos classificados como aquela classe.

**Revocação / Sensibilidade (Recall) por classe**:Fração de exemplos de uma classe corretamente identificados entre todos os exemplos reais daquela classe.

*F1-Score*:Média harmônica entre precisão e recall, útil quando há desbalanceamento de classes.

**Matriz de confusão (Confusion Matrix)**:Mostra quantos exemplos de cada classe foram confundidos com outras classes.

**Macro/ Weighted Averages de F1-Score, Precision, Recall**:
Permite avaliar o modelo levando em conta todas as classes simultaneamente.

3.3Como a base não é perfeitamente balanceada, a métrica mais indicada é o **F1-Score** , pois equilibra precisão e recall, evitando que o modelo pareça bom apenas por acertar as classes mais frequentes.
